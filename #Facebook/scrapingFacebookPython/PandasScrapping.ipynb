```python
import pandas as pd
#import matplotlib.pyplot as plt
import pylab
#%pylab inline
pylab.rcParams['figure.figsize'] = (30, 20)
#plt.rcParams["figure.figsize"] = 5000
```

```python
dataCs = pd.read_csv('Cs.Ciudadanos_facebook_statuses.csv', index_col=6)

dataCs.num_reactions.plot() # Vertex is the x-axis
pylab.xticks(size = 20)
pylab.yticks(size = 20)

pylab.xlabel('Fecha').set_fontsize(30)
pylab.ylabel('Número de reacciones a publicación').set_fontsize(40)
pylab.title('Reacciones a publicaciones de Ciudadanos').set_fontsize(40)
pylab.savefig('CsPosts.png')
```
