Check for null values

```python
df_statistics = []
for df in df_partidos:
    df_aux = pd.concat([pd.DataFrame.from_dict(item, orient='index').T for item in df.statistics]).dropna()
    df_statistics.append(df_aux)
# print(df_statistics[0])
```

```python
df_snippet = []
for df in df_partidos:
    df_aux = pd.concat([pd.DataFrame.from_dict(item, orient='index').T for item in df.snippet]).dropna()
    df_snippet.append(df_aux)
# print(df_snippet[0])
```

```python
df_topicDetails = []
for df in df_partidos:
    df_aux = pd.concat([pd.DataFrame.from_dict(item, orient='index').T for item in df.topicDetails.dropna()]).dropna()
    df_topicDetails.append(df_aux)
# print(df_topicDetails[0])
```

## Dislike Count Plot

```python
df = df_statistics[0]
print(df.describe(include='all'))
print(df.dtypes)
# print(df.commentCount.values.astype(float))
print(df.isnull().any())
print(df.head())
```

```python
rows = 3
cols = 3
fig, axs = plt.subplots(rows,cols)
t = 0;
for i in range(0,rows):
    for j in range(0,cols):
        if(t<len(df_statistics)):
            axs[i,j].plot(df_statistics[t]['dislikeCount'].values.astype(float))
            axs[i,j].set_title(names[t])
            t= t+1
            
fig.suptitle('Dislikes', fontsize=20)
```

```python
rows = 3
cols = 3
fig, axs = plt.subplots(rows,cols)
t = 0;
binwidth = 5
for i in range(0,rows):
    for j in range(0,cols):
        if(t<len(df_statistics)):
            data = df_statistics[t]['dislikeCount'].values.astype(float)
            axs[i,j].hist(data, bins=np.arange(min(data), 100 + binwidth, binwidth))
            axs[i,j].set_title(names[t])
            t= t+1
fig.suptitle('Dislikes', fontsize=20)
```
